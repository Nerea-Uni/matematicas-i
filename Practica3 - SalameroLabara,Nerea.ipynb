**Ejercicio 3**. Sea la serie $\displaystyle \sum_{k=1}^\infty\frac{1}{\sqrt{k}}$. Dibujar las sumas parciales de esta serie, $S_n=\displaystyle \sum_{k=1}^n\frac{1}{\sqrt{k}}$, y comprobar que son valores muy parecidos a $2\sqrt{n}$.

Continuación\*. La serie es divergente (hecho que ya sabíamos), pero lo anterior da información adicional: las sumas parciales divergen con la velocidad de $2\sqrt{n}$.

De hecho, se sabe demostrar que existe límite $\displaystyle \lim_{n\to\infty}(S_n-2\sqrt{n})=\zeta(1/2)\sim -1.46035$, donde $\zeta$ es la matemáticamente famosa función zeta de Riemann, que Sage tiene incorporada con la orden zeta(). Comprobar gráficamente que la sucesión $(S_n-2\sqrt{n})$ se va acercando a este número $\zeta(1/2)$.

```python
reset()
var('k')
L=[sum((1/sqrt(k)),k,1,a) for a in (1..50)]
list_plot(L)
```

```python
reset()
c=50
H=[0]
for n in [1..c]:
    H.append(H[n-1]+1/sqrt(n))
A=list_plot([H[n] for n in [1..c]],figsize=4)
B=list_plot([2*(sqrt(n)) for n in [1..n]],color='red',figsize=4)
A+B
```
